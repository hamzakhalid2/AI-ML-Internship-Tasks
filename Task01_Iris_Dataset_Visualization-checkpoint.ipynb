# Task 1: Exploring and Visualizing Iris Dataset

## Problem Statement
The Iris dataset contains measurements of iris flowers from three different species. Understanding the characteristics of the dataset through exploration and visualization helps identify patterns and relationships among the features.

## Goal
The goal of this task is to load the Iris dataset, explore its structure, and create visualizations to better understand feature distributions and relationships between variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Function: Load Dataset

This function loads the Iris dataset using Seaborn and returns it as a DataFrame for further analysis.

```python
def load_dataset():
    iris = sns.load_dataset("iris")
    return iris
```

## Function: Explore Dataset

This function displays important information about the dataset, including:
- Dataset shape
- Column names
- First five rows
- Data types and structure
- Summary statistics

```python
def explore_dataset(df):
    
    print("\n DATASET SHAPE ")
    print(df.shape)

    print("\n COLUMN NAMES ")
    print(df.columns.tolist())

    print("\n FIRST 5 ROWS ")
    print(df.head())

    print("\n DATASET INFORMATION ")
    df.info()

    print("\n SUMMARY STATISTICS ")
    print(df.describe())
```

## Function: Create Scatter Plot

This visualization shows the relationship between petal length and petal width for different iris species. Different colors represent different species.

```python
def plot_scatter(df):

    plt.figure(figsize=(8, 6))

    sns.scatterplot(
        data=df,
        x="petal_length",
        y="petal_width",
        hue="species"
    )

    plt.title("Petal Length vs Petal Width")
    plt.xlabel("Petal Length (cm)")
    plt.ylabel("Petal Width (cm)")
    plt.show()
```

## Function: Create Histograms

Histograms are used to visualize the distribution of each numerical feature in the dataset.

```python
def plot_histograms(df):

    df.hist(figsize=(10, 8))

    plt.suptitle("Distribution of Iris Features")
    plt.show()
```

## Function: Create Box Plots

Box plots help identify the spread of data and detect potential outliers in numerical features.

```python
def plot_boxplots(df):

    plt.figure(figsize=(10, 6))

    df.drop("species", axis=1).boxplot()

    plt.title("Box Plot of Iris Features")
    plt.ylabel("Measurement (cm)")
    plt.show()
```

## Main Function

The main function executes all steps of the analysis in sequence:
1. Load Dataset
2. Explore Dataset
3. Generate Scatter Plot
4. Generate Histograms
5. Generate Box Plots

```python
def main():

    # Step 1: Load Dataset
    iris = load_dataset()

    # Step 2: Explore Dataset
    explore_dataset(iris)

    # Step 3: Scatter Plot
    plot_scatter(iris)

    # Step 4: Histograms
    plot_histograms(iris)

    # Step 5: Box Plots
    plot_boxplots(iris)
```

## Program Execution

The following code runs the main function when the notebook is executed.

```python
if __name__ == "__main__":
    main()
```
