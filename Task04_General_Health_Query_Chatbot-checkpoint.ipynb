Task 4: General Health Query Chatbot

## Problem Statement
Many people have general health-related questions and need quick access to reliable information.

## Goal
Build a chatbot that answers general health questions using a Large Language Model (LLM) while ensuring responses are safe, friendly, and easy to understand.

```python
from transformers import pipeline
```

## Load the LLM

We use Microsoft's Phi-3 Mini model, which is freely available on Hugging Face.

```python
chatbot = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
```

## Safety Filter

This function blocks potentially dangerous medical requests.

```python
def safety_filter(user_query):

    dangerous_keywords = [
        "suicide",
        "overdose",
        "self harm",
        "stop my medicine",
        "prescribe medicine",
        "kill myself"
    ]

    query = user_query.lower()

    for keyword in dangerous_keywords:
        if keyword in query:
            return False

    return True
```

## Health Chatbot Function

Prompt engineering is used to make the chatbot act as a friendly medical assistant.

```python
def health_chatbot(user_query):

    if not safety_filter(user_query):
        return (
            "I cannot provide advice for potentially harmful "
            "medical situations. Please consult a qualified "
            "healthcare professional."
        )

    prompt = f"""
Act like a helpful medical assistant.

Rules:
- Be friendly and easy to understand.
- Provide general health information.
- Do not diagnose diseases.
- Do not prescribe medication.
- Encourage users to seek professional medical help for serious concerns.

Question:
{user_query}

Answer:
"""

    result = chatbot(
        prompt,
        max_new_tokens=200,
        do_sample=True,
        temperature=0.7
    )

    return result[0]["generated_text"]
```

## Test Example 1

```python
question = "What causes a sore throat?"

response = health_chatbot(question)

print(response)
```

## Test Example 2

```python
question = "Is paracetamol safe for children?"

response = health_chatbot(question)

print(response)
```

## Interactive Chatbot

```python
while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break

    response = health_chatbot(user_input)

    print("\nChatbot:")
    print(response)
    print()
```

## Results and Insights

- A health chatbot was successfully built using a Large Language Model.
- Prompt engineering was used to improve response quality and clarity.
- Safety filters were implemented to reduce harmful medical advice.
- The chatbot provides educational health information and encourages professional consultation when necessary.
